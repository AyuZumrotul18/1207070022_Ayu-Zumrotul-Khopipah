```python
#import library
import cv2 # Library OpenCV untuk pengolahan citra.
import numpy as np #Library untuk operasi array dan matriks.
from skimage import data #Library untuk mengakses data citra bawaan.
import matplotlib.pyplot as plt #Library untuk visualisasi data.
%matplotlib inline
```

```python
#Read Image
image = cv2.imread("anakgajah.jpg") # untuk membaca citra dengan nama file "anakgajah.jpg" dan menyimpannya dalam variabel image.
# Convert Image to Grayscale
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #untuk mengonversi citra dari format BGR (Blue-Green-Red) ke grayscale 

row, column = img.shape #untuk mendapatkan jumlah baris dan kolom citra. Nilai tersebut disimpan dalam variabel row (jumlah baris) dan column (jumlah kolom). 

img1 = np.zeros((row,column),dtype = 'uint8') #Membuat array kosong img1 dengan ukuran yang sama dengan citra img.
 

min_range = 10 #maka piksel pada citra img1 di set menjadi 255 (putih), jika tidak, maka di set menjadi 0 (hitam).
max_range = 60
 
#digunakan untuk mengiterasi setiap piksel dalam citra. 
for i in range(row): #Iterasi berdasarkan baris citra.
    for j in range(column): # Iterasi berdasarkan kolom citra.
        if img[i,j]>min_range and img[i,j]<max_range: #Jika nilai piksel pada citra img di posisi (i,j) lebih besar dari min_range dan lebih kecil dari max_range, maka kondisi terpenuh
            img1[i,j] = 255 #kondisi terpenuhi,
        else:
            img1[i,j] = 0 #kondisi tidak terpenuhi
```

```python
#plot image

fig, axes = plt.subplots(2, 2, figsize=(12, 12)) #untuk membuat grid dengan 2 baris dan 2 kolom untuk menampilkan citra dan histogram.
ax = axes.ravel() #untuk meratakan array dari objek axes menjadi 1 dimensi dan disimpan dalam variabel ax.

ax[0].imshow(img, cmap=plt.cm.gray) #untuk menampilkan citra img pada subplot pertama
ax[0].set_title("Citra Input") # untuk memberikan judul "Citra Input" pada subplot pertama.
ax[1].hist(img.ravel(), bins=256) #untuk menampilkan histogram citra img pada subplot kedua
ax[1].set_title('Histogram Input') #untuk memberikan judul "Histogram Input" pada subplot kedua.

ax[2].imshow(img1, cmap=plt.cm.gray) # untuk menampilkan citra img1 pada subplot ketiga
ax[2].set_title("Citra Output") #untuk memberikan judul "Citra Output" pada subplot ketiga.
ax[3].hist(img1.ravel(), bins=256) #untuk menampilkan histogram citra img1 pada subplot keempat 
ax[3].set_title('Histogram Output') # untuk memberikan judul "Histogram Output" pada subplot keempat.
```
