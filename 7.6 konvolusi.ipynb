```python
import matplotlib.pyplot as plt # untuk membuat visualisasi seperti plotting dan grafik.
%matplotlib inline

from skimage import data #modul dari library scikit-image yang menyediakan berbagai fungsi untuk pemrosesan dan manipulasi citra.
from skimage.io import imread # adalah fungsi dari skimage.io yang digunakan untuk membaca citra dari file.
from skimage.color import rgb2gray #untuk mengkonversi citra berwarna (RGB) menjadi citra grayscale.
import numpy as np # untuk melakukan operasi dan manipulasi array atau matriks dalam bahasa Python.

import cv2 # untuk pemrosesan citra dan vision komputer.
```

```python
citra1 = imread(fname="jerapah.tif") #untuk membaca citra file yang ditentukan oleh argumen 'fname'
print(citra1.shape) #mencetak dimensi citra dengan menggunakan atribut shape pada variabel 'citra1'

plt.imshow(citra1, cmap='gray') #untuk menampilkan gambar atau citra
```

```python
kernel = np.array([[-1, 0, -1], #Membuat kernel yang merupakan sebuah matriks 3x3 dengan nilai-nilai tertentu.
                   [0, 4, 0], #untuk melakukan operasi filtering pada 'citra1' menggunakan kernel yang telah dibuat
                   [-1, 0, -1]]) #

citraOutput = cv2.filter2D(citra1, -1, kernel)

fig, axes = plt.subplots(1, 2, figsize=(12, 12)) #Membuat sebuah figure dengan 1 baris dan 2 kolom subplot, dengan ukuran (12, 12).
ax = axes.ravel() #Mengubah array dari objek axes menjadi 1 dimensi dan disimpan dalam variabel `ax`.

ax[0].imshow(citra1, cmap = 'gray') #Menampilkan `citra1` pada subplot pertama (`ax[0]`) dengan menggunakan colormap 'gray'.
ax[0].set_title("Citra Input") # Memberikan judul "Citra Input" pada subplot pertama.
ax[1].imshow(citraOutput, cmap = 'gray') #Menampilkan `citraOutput` pada subplot kedua (`ax[1]`) dengan menggunakan colormap 'gray'.
ax[1].set_title("Citra Output")# Memberikan judul "Citra Output" pada subplot kedua.
```
